This notebook predicts temperature using a quadratic equation model:
T(t) = a * tÂ² + b * t + c

We'll go through 4 stages:
1. Hard-coded values
2. User input
3. File input (single set)
4. File input (multiple sets)


```python
# Stage 1: Hardcoded values
a = -0.01
b = 0.8
c = 10
t = 5

T = a * t**2 + b * t + c
print(f"[Stage 1] Predicted temperature at t={t} is {T:.2f}Â°C")
```

```python
# Stage 2: Input from user
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
t = float(input("Enter time t: "))

T = a * t**2 + b * t + c
print(f"[Stage 2] Predicted temperature at t={t} is {T:.2f}Â°C")
```

```python
# Stage 3: Read single set from file
with open("input.txt", "r") as file:
    a = float(file.readline())
    b = float(file.readline())
    c = float(file.readline())
    t = float(file.readline())

T = a * t**2 + b * t + c
print(f"[Stage 3] Predicted temperature at t={t} is {T:.2f}Â°C")
```

```python
# Stage 4: Multiple inputs from file
with open("input_multi.txt", "r") as file:
    lines = file.readlines()

for i, line in enumerate(lines, start=1):
    a, b, c, t = map(float, line.strip().split())
    T = a * t**2 + b * t + c
    print(f"[Set {i}] T({t}) = {T:.2f}Â°C")
```
